```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("✅ Libraries imported successfully!")
```

```python
x = np.linspace(0, 10, 200)
y = np.sin(x)

plt.plot(x, y)
plt.title("Simple Plot in Colab")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
X = data.data
y = data.target

print("✅ Dataset loaded!")
print("X shape:", X.shape)
print("y shape:", y.shape)
print("Classes:", data.target_names)
```

```python
df = pd.DataFrame(X, columns=data.feature_names)
df["target"] = y

df.head()
```

```python
print("Class distribution:")
print(df["target"].value_counts())

print("\nSummary statistics (first 5 features):")
df.describe().iloc[:, :5]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

print("Train set:", X_train.shape, y_train.shape)
print("Test set :", X_test.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=500, random_state=42)
model.fit(X_train_scaled, y_train)

print("✅ Model trained successfully!")
```

```python
y_pred = model.predict(X_test_scaled)

print("First 10 predictions:", y_pred[:10])
print("First 10 actual     :", y_test[:10])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay

acc = accuracy_score(y_test, y_pred)
print("✅ Test Accuracy:", acc)

cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", cm)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=data.target_names)
disp.plot()
plt.grid(False)
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=data.target_names))
```


**COMPARASION**

For a test_size of 0.25:

With StandardScaler: The accuracy is approximately 0.9860.
Without StandardScaler: The accuracy is approximately 0.9580.
As you can see, using the StandardScaler significantly improved the model's accuracy from 0.9580 to 0.9860. Scaling helps algorithms like Logistic Regression because it ensures that all features contribute equally to the distance calculations, preventing features with larger numerical ranges from disproportionately influencing the model and leading to faster and more stable convergence during training.
